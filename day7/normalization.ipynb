```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv("500hits.csv",encoding='latin-1')
```

```python
df
```

```python
df = df.drop(columns=['PLAYER', "CS"])
df
```

```python
df.info()
```

```python
df.describe()
```

```python
X1 = df.iloc[:,:-1]
```

```python
X2 = df.iloc[:,:-1]
```

```python
st = StandardScaler()
std = st.fit_transform(X1)
```

```python
mn = MinMaxScaler()
mn = mn.fit_transform(X2)
```

```python
df1= pd.DataFrame(std, columns=['YRS','G','AB','R','H','2B','3B','HR','RBI','BB','SO','SB','BA'])
```

```python
df1
```

```python
df2= pd.DataFrame(mn, columns=['YRS','G','AB','R','H','2B','3B','HR','RBI','BB','SO','SB','BA'])
```

```python
df2
```
