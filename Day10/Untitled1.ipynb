```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.DataFrame({"Data": [1, 2, 3 ,4 ,5 ,6 , np.nan, 8, 9, 10]})
df
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
mean_imp = SimpleImputer(strategy='mean')
```

```python
mean_imp.fit_transform(df)
```

```python
median_imp = SimpleImputer(strategy='median')
```

```python
median_imp.fit_transform(df)
```

```python
imp_mode = SimpleImputer(strategy="most_frequent")
```

```python
imp_mode.fit_transform(df)
```

```python
con_imp = SimpleImputer(strategy='constant', fill_value=7)
```

```python
con_imp.fit_transform(df)
```

```python
names = pd.DataFrame({'name':["vikram", 'dharun', 'dhina', np.nan, 'kishore']})
names
```

```python
imp_name = SimpleImputer(strategy = 'constant', fill_value='babe')
```

```python
imp_name.fit_transform(names)
```

```python
imp_mean_marked = SimpleImputer(strategy='mean', add_indicator=True)
```

```python
imp_mean_marked.fit_transform(df)
```

```python
df = pd.read_csv("my_csv.csv")
df
```

```python
from sklearn.compose import make_column_transformer
```

```python
ct =make_column_transformer(
                        (imp_name,['Name']),
                        (mean_imp,['miles']),
    remainder = 'drop'
)
```

```python
ct.set_output(transform='pandas')
```

```python
df = ct.fit_transform(df)
```
