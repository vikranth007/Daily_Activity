```python
import pandas as pd
```

```python
d = {'sales':[100000, 222000, 100000, 522000, 111111, 222222, 20000, 75000, 90000, 100000, 1000],
    'city': ["Tampa", "Tampa", "olfer", "dkaosd", "olfer", "dkaosd", "jibbi", "olfer","dkaosd", "olfer", "dkaosd"],
     'size':["small", "high", 'large', 'large', 'small', 'medium', 'high', 'medium', 'small', 'large', 'high']}
```

```python
df = pd.DataFrame(d)
df
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False)
```

```python
ode = OrdinalEncoder()
```

```python
from sklearn.compose import make_column_transformer
```

```python
ct = make_column_transformer(
    (ohe, ['city']),
    (ode,['size']),
    remainder = 'passthrough'
    
)
```

```python
ct.set_output(transform='pandas')
```

```python
df_pandas = ct.fit_transform(df)
```

```python
df_pandas
```

```python
# Drop
ct2 = make_column_transformer(
    (ohe, ['city']),
    (ode,['size']),
    remainder = 'drop'
    
)
```

```python
ct2.set_output(transform='pandas')
```

```python
df_pandas1 = ct2.fit_transform(df)
df_pandas1
```
