## Longest sub array with given sum -> positive and negative





```python
def longestsubarr(arr, k):
    n = len(arr)
    Sum = arr[0]

    left, right = 0 ,0 
    max_len = 0

    while right < n:
        while left <= right and Sum > k:
            Sum -=arr[left]
            left += 1

        if Sum == k:
            max_len = max(max_len, right - left + 1)

        right += 1

        if right < n:
            Sum += arr[right]
    return max_len

arr = [1, 2, 3, 4, 5]
k = 6
print(longestsubarr(arr, k))  # Output: 2
```

```python
def longest_sub_arr(arr, k):
    n = len(arr)
    prev_sum = 0
    max_len = 0
    seen = {}

    for i, num in enumerate(arr):
        prev_sum += num
        diff = prev_sum - k

        if prev_sum == k:
            max_len = i + 1

        if diff is seen:
             max_len = max(max_len, i - seen[diff])

        seen[diff] = i 

    return max_len         



arr = [1, 2, 3, 4, 5]
k = 6
print(longest_sub_arr(arr, k))
```
